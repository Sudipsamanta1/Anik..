# fillna

```python
df=pd.read_csv("/home/sudip/Desktop/dataset.csv")
```

```python
df.fillna(0) #filling 0 in position of missing value in the dataset 
```

```python
df.fillna({'name':"none",'Delhi':0})
```

```python
df.fillna(method='ffill') #fill the null values as forword(before) value in clm
```

```python
df.fillna(method='pad') #fill the null values as previous value in clm
```

```python
df.fillna(method='bfill') #fill the null values as backword(after) value in clm
```

```python
df.fillna(method='ffill',limit=2)
```
